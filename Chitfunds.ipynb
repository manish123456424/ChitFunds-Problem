```python
#calculating the actual contributed by the person
chit["Actual Contribution"] = chit["Contribution"] - chit["Amount returned to everyone in the group"]
chit.head()
```

```python
total=chit['Actual Contribution'].sum()
```

```python
total
#Hence the total amount is 43800
```

```python
chit["return"] = chit["Net amount recd by Bid winner"] - total
chit.head()
```

```python
#Return percentage = (Return / Total amount invested) X 100
chit["ret%"] = (chit["return"]/total)*100
chit.head()
```

```python
chit.tail()
```

```python
#What is the Annualized  Return of the person who bids in the last month ?
#25th month is the last month 
chit.iloc[24]
```

```python
#Formula for a person bids in last month
#((1 + return%)^(12/months the chit fund is hold) - 1) * 100

((1 + chit["ret%"][24]/100) ** (12/25) - 1) * 100


#Annualized  Return of the person who bids in the last month is 3.96%
```

```python
#What is the Annualized  Return of the person who bids in the first month ?
#1th month is the starting month

((1 + chit["ret%"][0]/100) ** (12/25) - 1) * 100

#Annualized  Return of the person who bids in the first month is -7.18%
```
